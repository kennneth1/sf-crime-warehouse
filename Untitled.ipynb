```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)   # show all columns, no truncation
pd.set_option('display.max_rows', 100)        # show more rows before truncating (adjust as needed)
pd.set_option('display.width', None)          # don't wrap wide tables
pd.set_option('display.max_colwidth', None)   # don't truncate long cell values (e.g. long text fields)
```

```python
df = pd.read_csv("sf_crime.csv")
```

```python
df.head(3)
```

```python
df = df[
    df['Incident Category'].notna() &
    df['Latitude'].notna() &
    df['Longitude'].notna() &
    df['Analysis Neighborhood'].notna() &
    df['Intersection'].notna() &
    (df['Police District'] != 'Out of SF')
]
```

```python
# Finalized resolutions
finalized = {
    "Cite or Arrest Adult",
    "Exceptional Adult",
    "Unfounded"
}

# Make sure report datetime is datetime
df["Report Datetime"] = pd.to_datetime(df["Report Datetime"])

# Higher priority for finalized resolutions
df["resolution_priority"] = df["Resolution"].isin(finalized).astype(int)

# Sort exactly like the SQL ORDER BY
df_sorted = df.sort_values(
    by=[
        "Incident Number",
        "Incident Code",
        "resolution_priority",
        "Report Datetime"
    ],
    ascending=[True, True, False, False]
)

# Equivalent of ROW_NUMBER()
df_sorted["rn"] = (
    df_sorted
    .groupby(["Incident Number", "Incident Code"])
    .cumcount() + 1
)

# Equivalent of WHERE rn = 1
df_latest = df_sorted[df_sorted["rn"] == 1]
```

```python
print(f"Original rows: {len(df):,}")
print(f"Latest offense rows: {len(df_latest):,}")
print(f"Rows removed: {len(df) - len(df_latest):,}")
print(f"Reduction: {(1 - len(df_latest)/len(df))*100:.2f}%")
```
