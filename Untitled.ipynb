```python
df = pd.read_csv("sf_crime.csv")
```

```python
df.head(3)
```

```python
filtered = df[
    df['Incident Category'].notna() &
    df['Latitude'].notna() &
    df['Longitude'].notna() &
    df['Analysis Neighborhood'].notna() &
    df['Intersection'].notna() &
    (df['Police District'] != 'Out of SF')
]
```

```python
filtered[['Intersection', 'Analysis Neighborhood', 'Police District']].drop_duplicates()
```

```python
print(filtered[['Intersection', 'Analysis Neighborhood', 'Police District']].drop_duplicates().shape[0])
print(filtered['Intersection'].nunique())
```

```python
print(filtered['Incident Datetime'].min(), filtered['Incident Datetime'].max())
```

```python
incident_counts = df.groupby("Incident Number").size()
```

```python
heavy_incidents = (
    df["Incident Number"]
    .value_counts()
    .loc[lambda x: x > 5]
    .index
)
```

```python
df_heavy = df[df["Incident Number"].isin(heavy_incidents)]
```

```python
df_heavy.sort_values(by="Incident Number")
```
